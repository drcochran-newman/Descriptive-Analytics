```python
# Color palettes for plots
# Named colors: https://matplotlib.org/stable/gallery/color/named_colors.html
# Seaborn color palette docs: https://seaborn.pydata.org/tutorial/color_palettes.html
# Seaborn palette chart: https://www.codecademy.com/article/seaborn-design-ii

# cp1 Color Palette - a binary blue/orange palette
blue = 'deepskyblue' # Use 'skyblue' for a lighter blue
orange = 'orange'
cp1 = [blue, orange]

# cp2 Palette - Reversed binary color order when needed for certain plots
cp2 = [orange, blue]

# cp5 Palette - 5 colors for use with categorical data
turqoise = 'mediumaquamarine'
salmon = 'darksalmon'
tan = 'tan'
gray = 'darkgray'
cp5 = [blue, turqoise, salmon, tan, gray]

# cpd Palette - blue-to-orange diverging palette for correlation heatmaps
cpd = sns.diverging_palette(242, 39, s=100, l=65, n=11)

# Set the default palette
sns.set_palette(cp1)
```

```python
# View cp1 color palette
sns.palplot(cp1)
```

```python
# View cp2 color palette
sns.palplot(cp2)
```

```python
# View cp5 color palette
sns.palplot(cp5)
```

```python
# View Diverging color palette cpd
sns.palplot(cpd)
```

# Example Plots Decluttered and with New Color Palettes

## Histograms
See the docs: https://seaborn.pydata.org/generated/seaborn.histplot.html

```python
# Basic overlaid histogram with Survived as target variable
# Using the cpbinary custom palette, which has been set as default
sns.histplot(data=df, x='Age', hue='Survived');
```

```python
# Colors reversed to show minority class in front
# Using the cpbinary custom palette, which has been set as default
sns.histplot(data=df, x='Age', hue='Survived', hue_order=[1,0], palette=cp2);
```

### Enhanced histogram
- adjusted plot with the line: `fig = plt.figure(figsize=(12,6))`
- specified bin widths with `binwidth=`
- adjust transparency with `alpha=` (decimal between 0 and 1)

```python
# Enhanced histogram for Age with figure sizing, title, reduced alpha-transparency, and optional line to customize axis labels
# Colors reversed to show minority class in front
# Using the cpbinary custom palette, which has been set as default
plt.figure(figsize=(8,4))
plt.title("Age Distributions Comparison", fontsize=14, fontweight='bold')
ax = sns.histplot(data=df, x='Age', hue='Survived', hue_order=[1,0], palette=cp2, binwidth=5, alpha=0.7);
# ax.set(xlabel = 'Custom x axis label', ylabel='Custom y axis label');
```

## Box Plots

See the docs: https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
# Basic Boxplot comparing distributions of Age for target variable Survived
sns.boxplot(data=df, x='Survived', y='Age');
```

**Enhanced Horizontal Boxplot**
- Set dimensions with the fig variable
- Turn horizontal by switching the x and y variables and adding `orient='h'`
- Added transparency, which for boxplots requires adding a slightly more complex "boxprops" parameter. See this discussion: https://github.com/mwaskom/seaborn/issues/979

```python
# Enhanced HORIZONTAL Boxplot, with figure sizing, title, and optional axes labels
plt.figure(figsize=(8,4))
plt.title("Age Distributions Comparison", fontsize=14, fontweight='bold')
ax = sns.boxplot(data=df, x='Age', y='Survived', orient='h');
# ax.set(xlabel = 'Custom x axis label', ylabel='Custom y axis label');
```

```python
# Violin Plot
plt.title("Age Distributions Comparison", fontsize=14, fontweight='bold')
ax = sns.violinplot(data=df, x='Age', y='Survived', orient='h');
# ax.set(xlabel = 'Custom x axis label', ylabel='Custom y axis label');
```

```python
# Violin Plot
# Scaled by count: scale='count'
plt.title("Age Distributions Comparison", fontsize=14, fontweight='bold')
ax = sns.violinplot(data=df, x='Age', y='Survived', orient='h', scale='count');
# ax.set(xlabel = 'Custom x axis label', ylabel='Custom y axis label');
```

```python
# Violin Plot - split by binary class
# blank variable allows to split colors by binary class in a single violin plot
df[all] = " "
plt.title("Age Distributions Comparison", fontsize=14, fontweight='bold')
ax = sns.violinplot(data=df, x='Age', y=all, hue='Survived', split=True);
# ax.set(xlabel = 'Custom x axis label', ylabel='Custom y axis label');
```

```python
# Violin Plot - split and scaled
# blank variable allows to split colors by binary class in a single violin plot
# scale="count" scales its plot by its count of records
df[all] = " "
plt.title("Age Distributions Comparison", fontsize=14, fontweight='bold')
ax = sns.violinplot(data=df, x='Age', y=all, hue='Survived', split=True, scale='count');
# ax.set(xlabel = 'Custom x axis label', ylabel='Custom y axis label');
```

## Count Plots
- A specialized bar plot that gives a simple count of records per category (or ordinal).
- See the docs: https://seaborn.pydata.org/generated/seaborn.countplot.html#seaborn.countplot

```python
# Countplot
plt.title("Male / Female Survival Comparison", fontsize=14, fontweight='bold')
ax = sns.countplot(data=df, x='Sex', hue='Survived');
# ax.set(xlabel = 'Custom x axis label', ylabel='Custom y axis label');
```

**Countplots can also be used with ordinal variables.**

```python
# Compare survivors vs. non-survivors by Pclass
plt.title("Pclass Survival Comparisons", fontsize=14, fontweight='bold')
ax = sns.countplot(data=df, x='Pclass', hue='Survived');
# ax.set(xlabel = 'Custom x axis label', ylabel='Custom y axis label');
```

## Bar Plots
- For comparing categorical variables and/or or ordinal variables.
- The Seaborn barplot is designed to compare the *mean* for multiple categories.
- By default, the box plot uses a separate color per category and includes a confidence interval mark, which looks like a candle wick.
- See the docs: https://seaborn.pydata.org/generated/seaborn.barplot.html

### Enhanced Barplot
- Normally we need only one color, which we can specify using the `color` parameter.
- Remove the confidence interval mark with `ci=None`

```python
# Compare average survival rate by SibSp using blue for one color
plt.title("SibSp Survival Rates", fontsize=14, fontweight='bold')
ax = sns.barplot(data=df, x='SibSp', y='Survived', ci=None, color=blue);
# ax.set(xlabel = 'Custom x axis label', ylabel='Custom y axis label');
```

### Bar Plot using color with purpose
The barplot below comparing the proportion of men and women per category.

```python
# Compare average survival of Men and Women by SibSp
# Note that this uses our cp2 color palette defined above
plt.title("Survival of Men vs. Women by SibSp", fontsize=14, fontweight='bold')
ax = sns.barplot(data=df, x='SibSp', y='Survived', palette=cp2, ci=None, hue='Sex');
# ax.set(xlabel = 'Custom x axis label', ylabel='Custom y axis label');
```

## Correlation Heatmap
- https://seaborn.pydata.org/generated/seaborn.heatmap.html
- https://medium.com/@szabo.bibor/how-to-create-a-seaborn-correlation-heatmap-in-python-834c0686b88e
- https://www.geeksforgeeks.org/how-to-create-a-triangle-correlation-heatmap-in-seaborn-python/

```python
# Create a subset of numeric variables
cont=['Age','Fare','Pclass','Parch','SibSp']
df[cont].head()
```

```python
# Get a numeric report of Pearson correlations ranging from -1 to 1
df[cont].corr()
```

## Seaborn Default Heatmap

```python
# Seaborn's default heatmap colors used to plot correlation as heatmap
# Hide the upper right repetitive correlations
mask = np.triu(np.ones_like(df[cont].corr()))
# Display a default heatmap with labels
ax = sns.heatmap(df[cont].corr(), vmin=-1, vmax=1, mask=mask, annot=True);
```

## Enhanced Correlation Heatmap
This uses the custom `cp3` diverging color palette defined above, using the `cmap` parameter.

```python
# Customized heatmap with our custom diverging color palette defined above
# Hide the upper right repetitive correlations
mask = np.triu(np.ones_like(df[cont].corr()))
plt.figure(figsize=(7,4))
plt.title("Correlation Heatmap", fontsize=14, fontweight='bold')
ax = sns.heatmap(df[cont].corr(), cmap=cp3, vmin=-1, vmax=1, mask=mask, annot=True);
```
